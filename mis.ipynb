```python
import matplotlib.pyplot as plt
import numpy as np



# Create the figure and axes
fig, ax = plt.subplots()
# Data for the random walk
x_random = np.linspace(0, 11, 10)
y_randoms = []

colors = ['mediumseagreen', 'skyblue', 'brown']

for i in range(3):
# Plot the random walk
    y_random = np.random.rand(10) * 1.5 + 0.8
    y_randoms.append(y_random)
    ax.plot(x_random, y_random + i*1.5, color=colors[i], label='Random Walk', marker='o')
    ax.scatter([11.4, 11.7, 12], [y_randoms[i][-1]+ i*1.5]*3, color=colors[i])
    

ax.set_ylim(0, 6)
ax.set_axis_off()
```
